```python
import requests
#import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
#import time
#import pandas as pd
```

```python
url="https://www.naukri.com/data-science-recruiters-in-hyderabad-secunderabad"
page=requests.get(url)
soup=bs(page.text,'html.parser')
print(soup.prettify())
```

```python
#job_title


def Extract_hr_name(soup):
    hrname=[]
    spans=soup.findAll(name='span',attrs={'class':'fl ellipsis'})
    for span in spans:
        hrname.append(span.text.strip())
    return(hrname)
    

Extract_hr_name(soup)


# def Extract_Comapny_Locations_result(soup):
#     hr=[]
#     spans=soup.findAll('span',attrs={'class':'fl ellipsis'})
    
#     for span in spans:
        
#         hr.append(span.text)
#     return(hr)
# Extract_Comapny_Locations_result(soup)


# def Extracting_skils_hirefor(soup):
#     hire=[]
#     div=soup.findAll(name='small',attrs={'class':'ellipsis'})
#     for d in div:
#         hire.append(d.text.strip())
#     return(hire)
        
# Extracting_skils_hirefor(soup)


def Extracting_location_hirefor(soup):
    hire=[]
    div=soup.findAll(name='small',attrs={'class':'ellipsis'})
    for d in div:
        hire.append(d.text.strip())
    return(hire)
        
Extracting_location_hirefor(soup)



def Extracting_skils_hirefor(soup):
    hire=[]
    div=soup.findAll(name='div',attrs={'class':'hireSec highlightable'})
    for d in div:
        hire.append(d.text.strip())
    return(hire)
        
Extracting_skils_hirefor(soup)

#Lats Active Date
def lats_active_days(soup):
    #Active date(a_date) last seen date
    a_date=[]
    spans=soup.findAll(name='div',attrs={'class':'activeJobs'})
    for span in spans:
        a_date.append(span.text.strip())
    return(a_date)
lats_active_days(soup)


'''=================================>All Veriables Assing<==========================================='''





Hr_Names=Extract_hr_name(soup)
Company_location=Extracting_location_hirefor(soup)
last_active=lats_active_days(soup)

Hiring_skills=Extracting_skils_hirefor(soup)
columns={'Hr_Names':Hr_Names,'Company_location':Company_location,'Hiring_skills':Hiring_skills,'Last_Active_Date':last_active}

df=pd.DataFrame.from_dict(columns,orient='index')
df=df.transpose()
df.to_csv(r'C:\\Users\\kumar\Desktop\\WebScraping_Data\\NaukarData01.csv', index = False)

df




    
```
